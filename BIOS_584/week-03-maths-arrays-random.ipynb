<font size = "4">
Create an array from a list

$ \text{ }$
$ a = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}
\qquad b = \begin{pmatrix} 0 \\ 1 \\ 0\end{pmatrix}
\qquad c = \begin{pmatrix} 1 \\ 10 \\ 100 \\ 1000 \\ 10000 \end{pmatrix} 
\qquad d = \begin{pmatrix} 4 \\ 2 \end{pmatrix} $

```python
# You create an array by converting a list [1,2,3] to a numeric object
# "array" is a subfunction of the library numPy.

vec_a  = np.array([1,2,3])
vec_b  = np.array([0,1,0])
print(vec_a)
print(vec_b)
```

<font size = "4">

Accessing an element of an array

```python
# Arrays, like lists, start their numbering at zero.
# You also access them with square brackets

print(vec_a[0])
print(vec_a[2])
```

<font size = "4">
Operations with a single array and a scalar

<font size = "4">

$ a + 3 = \begin{pmatrix} a_1 + 3 \\ a_2 + 3 \\ a_3 + 3 \end{pmatrix}$

```python
print(vec_a)
```

```python
# A scalar refers to either an int or float
# We can do many common operations with 

print(vec_a * 3)
print(vec_a / 3)
print(vec_a + 3)
print(vec_a ** 3)
```

<font size = "5">
Element-by-element addition between two arrays of the same size

$\text{ }$

$ a + b = \begin{pmatrix} a_1 \\ a_2 \\ a_3 \end{pmatrix} +
          \begin{pmatrix} b_1 \\ b_2 \\ b_3 \end{pmatrix} =
          \begin{pmatrix} a_1 + b_1 \\ a_2 + b_2 \\ a_3 + b_3 \end{pmatrix}$

```python
print(vec_a)
print(vec_b)
```

```python
# When you add two arrays of the same size,
# Python adds the individual elements 
# in each position

print(vec_a + vec_b)
```

<font size ="5">

The default multiplication option is element-by element:

$ a * b = \begin{pmatrix} a_1 * b_1 \\ a_2 * b_2 \\ a_3 * b_3 \end{pmatrix}$

```python
# We can do other similar element-by-element operations
# such as multiplication, subtraction, and division.

print(vec_a - vec_b)
print(vec_a * vec_b)
print(vec_a / vec_b)
```

<font size = "4">
Summary statistics of an array

```python
# The function np.mean() computes the mean of an array
# We can do similar operations for the standard deviation (std),
# the minimum (min), the median (median), and the maximum (max).
print(vec_a)
print(np.mean(vec_a))
print(np.std(vec_a))
print(np.min(vec_a))
print(np.median(vec_a))
print(np.max(vec_a))
print(np.cumsum(vec_a))
print(np.cumprod(vec_a))
```

## 4. Random Numbers

<font size = "4">

Why randomness?

- Simulate different scenarios: high risk vs low risk; different patient subpopulation
- Study proportions of a complex system and/or estimator
- Randomly assign subjects to treatment or control in clinical trials
- ...

<font size = "4">

Create a vector of random variables

```python
import random
```

```python
# This code creates a vector of random variables 
# generated from a normal distribution with
# mean "loc" (location) and standard deviation "scale"
# the number of distinct variabels is "size"

randomvar_a = np.random.normal(loc=0, scale=1, size=10)
print(randomvar_a)
```

<font size = "4">

Random numbers differ every time!

- Avoid this problem by drawing from a "pregenerated" set.
- This is called a seed

```python
# If you set the seed, you can replicate your simulation results
# Try running this code section multiple times and see if the random_var_b changes?

np.random.seed(612)

random_var_b = np.random.normal(loc=0, scale=1, size=10)
print(random_var_b)
```

<font size = "4">

Compute a histogram with the results

```python
# Compute a histogram
np.random.seed(612)
randomvar_x = np.random.normal(loc=0, scale=1, size=100)

plt.hist(x = randomvar_x)
plt.xlabel("Variable x")
plt.ylabel("Frequency")
plt.title("Histogram Plot")
plt.show()
```

```python
mean_val = np.mean(randomvar_x)
sd_val = np.std(randomvar_x)
print(mean_val)
print(sd_val)
```
